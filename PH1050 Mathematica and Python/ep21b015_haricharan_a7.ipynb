## Create Riemann sum function including all three methods (left, right and mid-point) and test the function on f(x) for different values of N = 10, 100, 1000...for all 3 methods (Take a = 0, b = 5). You should evaluate the integral in each method upto that value of N for which an accuracy goal of 3 decimal places has been reached. Accuracy goal is considered to be fulfilled when going to the next order of N doesn’t change the answer. Print all the values of integral for each method in a list. Also, print the value of minimum N (in the order of 10) which is required to reach the accuracy goal.

```python
def riemann_sum(function, a, b, N, method):
    #Defining dx
    dx = (b - a)/N
    #Creating an array of x values
    x = np.linspace(a,b,N+1)
    if method == 'left':
        x_left = x[:-1]
        return np.sum(function(x_left)*dx)
    elif method == 'right':
        x_right = x[1:]
        return np.sum(function(x_right)*dx)
    elif method == 'midpoint':
        x_mid = (x[:-1] + x[1:])/2
        return np.sum(function(x_mid)*dx)
    raise Error("\"method\" must be left, right or midpoint")
```

```python
def f(x):
    return (x**3 + 4)/(x**2 - 2 * np.sin(x) + 3)
```

```python
#To find required N value
def test_and_print(method):
    print("\n", method.upper(), sep = "")
    N = 10
    previous_sum = np.infty
    current_sum = riemann_sum(f, 0, 5, N, method)
    
    while np.round(previous_sum, 3) != np.round(current_sum, 3):
        print(current_sum)
        N*= 10
        previous_sum = current_sum
        riemann_sum_value = riemann_sum(f, 0, 5, N, method)
        current_sum = riemann_sum_value
    print("Minimum N is: ", N)
test_and_print('left')
test_and_print('right')
test_and_print('midpoint')
```

## 2. Suppose a rod of length 10 meter has mass density ρ(x)= kx, where k is the proportionality constant and x is the distance from one end of the rod. Calculate total mass of the Rod using trapezoid rule. (Take k = 2). Compare this with Riemann sum.

```python
def trapezoidal_rule(function, a, b, N):
    return (riemann_sum(function, a, b, N, 'left') + riemann_sum(function, a, b, N, 'right'))/2
```

```python
k = 2
def rho(x):
    return k*x
```

```python
N = 10000
riemann_sum_value = riemann_sum(rho, 0, 10, N, 'left')
trapezoidal_sum_value = trapezoidal_rule(rho, 0, 10, N)
print("riemann_sum_value = ", riemann_sum_value, "\ntrapezoidal_sum_value = ", trapezoidal_sum_value)
```

## 3. Find the center of mass of a 2-meter rod lying on the x − axis with its left end at the origin if its density is ρ(x) = 15x^2 kg/m. Use Simpson’s rule for this and compare the answer with Riemann sum.

```python
def mass_density(x):
    return 15 * (x**2)
```

```python
def simpson_sum(function, a, b, N):
  midpoint = riemann_sum(function, a, b, N//2, "midpoint")
  trapezoidal = trapezoidal_rule(function, a, b, N//2)
  return ((2 * midpoint + trapezoidal)/3)
```

```python
N = 1000
riemann_sum_value = riemann_sum(mass_density, 0, 2, N, 'midpoint')
simpson_sum_value = simpson_sum(mass_density, 0, 2, N)
print("riemann_midpoint_sum: ", riemann_sum_value, "\nsimpson_sum_value", simpson_sum_value)
```

### a) $10^4$ nuclides of Francium-223 (F r223) (with a half-life of 22 minutes) decay into Radium-223over the course of 3 hours. Solve the radioactive decay equation using Euler method and plot the number of nuclides (NA) as a function of time in a log-log plot.

```python
#Note: all units are in minutes
decay_constant_a = np.log2(2)/22 
def dNdt(N, t):
    return -1 * decay_constant_a * N
t_values = np.arange(0, 181, 1)
y_values = odeint(dNdt, 10000, t_values)
```

```python
plt.loglog(t_values, y_values)
plt.xlabel("time (in minutes)")
plt.ylabel("$N_A$")
plt.title("Number of nuclides of $Fr^{223}$")
plt.legend(["Number of nuclides"],
           fontsize = "x-small")
plt.show()
```

The LotkaVolterra equations (or the predatorprey equations) are frequently used to describe the dynamics of biological systems in which two species interact, one as a predator and the other as prey. Let x denote the deer population and y denote the lion population in a forest. The population dynamics over time are given by:

## where a, b, c, f are parameters which describe the interaction between predator and prey.Solve the following questions using odeint from scipy package:

```python
#Vector v = [x, y]
def dvdx(v, t):
    x = v[0]
    y = v[1]
    return np.array([a * x - b * x * y, c * x * y - f * y])
```

### (a) Assuming the initial conditions x(0) = 1.5, y(0) = 1, plot the deer and lion population as a function of time, for t ∈ (0, 10) in the same plot for each of these interaction parameters (total 3 plots with 2 curves each), and put the values of interaction parameters as the title of the plot.

```python
v0 = [1.5, 1]
t_values = np.arange(0, 10, 0.1)
for a, b, c, f in [1,1,1,1], [0.5, 0.5, 1, 1], [0.25, 1, 0.5, 0.75]:
    v_values = odeint(dvdx, v0, t_values)
    plt.plot(t_values, v_values[:, 0])
    plt.plot(t_values, v_values[:, 1])
    plt.xlabel("time")
    plt.ylabel("Population")
    plt.title(f"a = {a}, b = {b}, c = {c}, f = {f}")
    plt.legend(["Deer population", "Lion Population"],
           fontsize = "x-small")
    plt.show()
```
