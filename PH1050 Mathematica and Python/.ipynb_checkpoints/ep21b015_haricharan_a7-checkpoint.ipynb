## Create Riemann sum function including all three methods (left, right and mid-point) and test the function on f(x) for different values of N = 10, 100, 1000...for all 3 methods (Take a = 0, b = 5). You should evaluate the integral in each method upto that value of N for which an accuracy goal of 3 decimal places has been reached. Also, print the value of minimum N (in the order of 10) which is required to reach the accuracy goal.

Defining a riemann sum function, that takes in a function, a, b, N, and method as the input and returns the appropriate riemann sum

```python
def riemann_sum(function, a, b, N, method):
    #Defining dx to be width of the interval
    dx = (b - a)/N

    #Creating an array of x values
    x = np.linspace(a,b,N+1)
    
    #Checking methods
    if method == 'left':
        x_left = x[:-1]
        return np.sum(function(x_left)*dx)
    elif method == 'right':
        x_right = x[1:]
        return np.sum(function(x_right)*dx)
    elif method == 'midpoint':
        x_mid = (x[:-1] + x[1:])/2
        return np.sum(function(x_mid)*dx)
    raise Error("\"method\" must be left, right or midpoint")
```

Defining the required function

```python
def f(x):
    return (x**3 + 4)/(x**2 - 2 * np.sin(x) + 3)
```

Printing all of the required values for the required methods.

```python
#To find required N value
def test_and_print(method):
    print("\n", method.upper(), sep = "")
    #Initiializing N to be 10
    N = 10
    #Initializing previous_sum to be a very large number, i.e. infinity
    previous_sum = np.infty
    #Initializing current_sum to be the riemann sum
    current_sum = riemann_sum(f, 0, 5, N, method)
    
    #Using a while loop till the condition is reached
    while str(previous_sum)[:6] != str(current_sum)[:6]:
        print(current_sum)
        N*= 10
        previous_sum = current_sum
        riemann_sum_value = riemann_sum(f, 0, 5, N, method)
        current_sum = riemann_sum_value    
    print("Minimum N is: ", N)

#Using the test_and_print function to print outputs for various methods
test_and_print('left')
test_and_print('right')
test_and_print('midpoint')
```

## 2. Suppose a rod of length 10 meter has mass density ρ(x)= kx, where k is the proportionality constant and x is the distance from one end of the rod. Calculate total mass of the Rod using trapezoid rule. (Take k = 2). Compare this with Riemann sum.

Defining a trapezoidal rule function

```python
#Defining a trapezoidal rule function
def trapezoidal_rule(function, a, b, N):
    #Defining dx to be width of integration
    dx = (b - a)/N

    #Creating an array of x values
    x = np.linspace(a,b,N+1)
    
    return (function(x[0])+2*np.sum(function(x[1:-1])) + function(x[-1]))*dx/2
```

Defining the required $\rho$ function

```python
k = 2
def rho(x):
    return k*x
```

Outputting the required riemann and trapezoidal sums

```python
N = 1000
riemann_sum_value = riemann_sum(rho, 0, 10, N, 'left')
trapezoidal_sum_value = trapezoidal_rule(rho, 0, 10, N)

print("riemann_sum_value = ", riemann_sum_value, "\ntrapezoidal_sum_value = ", trapezoidal_sum_value)
```

Clearly, the trapezoidal sum is more accurate!

## 3. Find the center of mass of a 2-meter rod lying on the x − axis with its left end at the origin if its density is ρ(x) = 15x^2 kg/m. Use Simpson’s rule for this and compare the answer with Riemann sum.

Defining the simpson_sum function

According to the simpson_sum expression, if there are 2n subdivisions, $\text{Integral} \approx \frac {\Delta x}{6} (f(a) + 4 f(a + \frac{\Delta x}2) + 2 f(a + \Delta x) + 4 (f(a + \frac{3 \Delta x}{2})) + \dots + f(b))$.

With some algebra, Simpson sum with N terms = $\frac23 \text{(Midpoint sum with N/2 terms)} + \frac 13  \text{(Trapezoidal Sum with N/2 terms)}$

```python
def simpson_sum(function, a, b, N):
  midpoint = riemann_sum(function, a, b, N//2, "midpoint")
  trapezoidal = trapezoidal_rule(function, a, b, N//2)
  return ((2 * midpoint + trapezoidal)/3)
```

Defining the required mass density function

```python
def mass_density(x):
    return 15 * (x**2)
```

Outputting the required values

```python
N = 1000
riemann_sum_value = riemann_sum(mass_density, 0, 2, N, 'left')
simpson_sum_value = simpson_sum(mass_density, 0, 2, N)
print("riemann_midpoint_sum: ", riemann_sum_value, "\nsimpson_sum_value", simpson_sum_value)
```

The simpson sum is more accurate than the Riemann sum!

Defining the decay constant as per the formula and defining dNdt function

```python
#Note: all units are in minutes
decay_constant_a = np.log2(2)/22 
def dNdt(N, t):
    return -1 * decay_constant_a * N
t_values = np.arange(0, 181, 1)
y_values = odeint(dNdt, 10000, t_values)
```

Plotting everything using matplotlib

```python
#Using a log log plot
plt.loglog(t_values, y_values)

#Setting x and y labels
plt.xlabel("time (in minutes)")
plt.ylabel("$N_A$")

#Giving a title to the plot
plt.title("Number of nuclides of $Fr^{223}$")
plt.legend(["Number of nuclides"],
           fontsize = "x-small")
plt.show()
```

Defining two decay constants. The differential equation is solved as a vector differential equation, with the vector $\begin{bmatrix} N_A \\ N_B\end{bmatrix}$

```python
#Consider a vector of [N_A, N_B]
#Using hours as unit
decay_constant_a = np.log2(2)/(22/60) 
decay_constant_b = np.log2(2)/(11.4 * 24) 
N_A0 = 10000

#Defining a dvdt, i.e. the derivative of the vector
def dvdt(vector, t):
    return np.array([- decay_constant_a * vector[0], 
         -decay_constant_b * vector[1] + decay_constant_a * N_A0 * np.exp(- decay_constant_a * t)])
```

Using the odeint function to find the required values, and subsequently plotting them using matplotlib

```python
#Defining a t_values array
t_values = np.arange(0, 24*10, 0.5)
vector_values = odeint(dvdt, [N_A0, 0], t_values)

#Using a log-log plot
plt.loglog(t_values, vector_values[:, 1])
#Setting y limits
plt.ylim(1, 2*10**4)
#Setting axes labels, and plot titles
plt.xlabel("time (in hours)")
plt.ylabel("$N_B$")
plt.title("Number of nuclides of $Ra^{223}$")
plt.legend(["Number of nuclides"],
           fontsize = "x-small")
plt.show()
```

The LotkaVolterra equations (or the predatorprey equations) are frequently used to describe the dynamics of biological systems in which two species interact, one as a predator and the other as prey. ## where a, b, c, f are parameters which describe the interaction between predator and prey.Solve the following questions using odeint from scipy package:

Defining a vector, and finding the vector derivative function

```python
#Vector v = [x, y]
def dvdt(v, t):
    x = v[0]
    y = v[1]
    return np.array([a * x - b * x * y, c * x * y - f * y])
```

### (a) Assuming the initial conditions x(0) = 1.5, y(0) = 1, plot the deer and lion population as a function of time, for t ∈ (0, 10) in the same plot for each of these interaction parameters (total 3 plots with 2 curves each), and put the values of interaction parameters as the title of the plot.

For the given initial values, plotting the deer and lion population as a function of time, with the interaction parameters as the title of the plot

```python
#Initial conditions vector
v0 = [1.5, 1]
#Defining t values numpy array
t_values = np.arange(0, 10, 0.1)
for a, b, c, f in [1,1,1,1], [0.5, 0.5, 1, 1], [0.25, 1, 0.5, 0.75]:
    v_values = odeint(dvdt, v0, t_values)
    
    #Plotting lion and deer populations
    plt.plot(t_values, v_values[:, 0])
    plt.plot(t_values, v_values[:, 1])
    
    #Setting x and y labels
    plt.xlabel("time")
    plt.ylabel("Population")
    plt.title(f"a = {a}, b = {b}, c = {c}, f = {f}")
    plt.legend(["Deer population", "Lion Population"],
           fontsize = "x-small")
    plt.show()
```
